### fill_between(x,y1,y2)

```python
# theme
mystyle = plt.style.available[2]
plt.style.use(mystyle)
# make data
np.random.seed(1)
x = np.linspace(0, 8, 16) # get evenly spaced numbers in a given interval
y1 = 3 + 4*x/8 + np.random.uniform(0, 0.5, len(x))  # value for upper y, scale down by x (see graph)
y2 = 1 + 2*x/8 + np.random.uniform(0, 0.5, len(x))  # value for lower y, scale down by x (see graph)

# plot
fig, ax = plt.subplots() 
ax.fill_between(x, y1, y2, alpha=0.4, linewidth=0)  # fill between 
ax.plot(x, (y1 + y2)/2, linewidth=2)  # line plot of the average between y1 and y2
ax.set(xlim=(0,8), xticks=np.arange(1,8),
       ylim=(0,8), yticks=np.arange(1,8))  # set axes limits and ticks
plt.title('style: '+ mystyle)
plt.show()  # show plot
```

### streamplot(X,Y,U,V)
- like a stream 
- a streamlines of a vector flow
- need more study on [streamplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.streamplot.html#matplotlib.axes.Axes.streamplot) and its use cases

```python
# make data:
# make stream function
X, Y = np.meshgrid(np.linspace(-3,3,256), np.linspace(-3,3,256))  
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of stream function
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V, color='green')
plt.show()
```

```python
# Playing around
X, Y = np.meshgrid(list(range(10)), list(range(10)))
print('X: \n', X)
print('Y: \n', Y)
Z = X**2 + Y**2
print('Z: \n', Z)
V = np.diff(Z[1:, :], axis=1)  # Calculate the n-th discrete difference along the given axis.
print('Z[1:, :] \n', Z[1:, :])
print('V: \n', V)
U = -np.diff(Z[:, 1:], axis=0)
print('Z[:, 1:] \n', Z[:, 1:])
print('U: \n', U)
fig, ax = plt.subplots()
ax.streamplot(X[1:, 1:],Y[1:, 1:],U,V)
plt.show()

# slicing a list
a = list(range(10))
print('a: ', a)
print('a[1:]', a[1:])
print('a[:1]', a[:1])
```

### violinplot(D)
- Violin Plots are used when we want to see the distribution of Numerical Data 
- specially useful when we want to make a comparison of distributions between multiple groups 
- find more here [YouTube](https://www.youtube.com/watch?v=ktZH1OUvD4o&ab_channel=GraphPadSoftware)

```python
# make data
np.random.seed(10)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))

# plot
fig, ax = plt.subplots()

vp = ax.violinplot(D, [2, 4, 6], widths=2,
                   showmeans=True, showmedians=True, showextrema=True)

# styling
for body in vp['bodies']:
  body.set_alpha(0.9)

ax.set(xlim=(0,8), xticks=np.arange(1,8),
       ylim=(0,8), yticks=np.arange(1,8))

plt.show()
```

### hist2d(x,y)
Usage of a 2D Histogram: 
- analyse the relationship among two data variables which have a wide range of values. 
- a 2d histogram is very similar to 1d histogram in concept but looks quite different
  - unline 1D histogram, a 2D histogram is drawn by including the total number of combinations of the values which occur in intervals of x and y, and marking the densities
- it is useful when there is a large amount of data in a discrete distribution
- it simplifies the data by visualising the points where the frequencies of the variables are dense

- in **image processing** a 2d histogram shows the relstionship of intensities at the exact position between two images

```python
# make data
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

print(x[:5])
print(y[:5])

# plot
fig, ax = plt.subplots()
ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)), cmap='Blues')
ax.set(xlim=(-2,2), ylim=(-3,3))

plt.show()
```

```python
# Understanding the above
from matplotlib import cm

# a very simple 2d histogram
x = [1,2,3,4,5,6,7,8,9]
y = [1,3,5,7,9,11,13,15,17]
plt.hist2d(x,y, cmap='Blues')
plt.show()

# example 2 - with more matches
x = [1,2,3,4,5,6,7,8,9]
y = [1,2,2,3,4,4,4,8,8]
np.random.shuffle(x)
np.random.shuffle(y)
plt.hist2d(x,y, cmap='BuPu')
plt.show()

# example 3
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
plt.hist2d(x, y, cmap=cm.gist_rainbow)
plt.show()

# example 4 increasing the number of bins
plt.hist2d(x, y, bins=(100, 100), cmap=cm.gist_heat_r)
plt.colorbar()
plt.show()
```
