#### Your first plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0, 2* np.pi, 200)  # generate 200 evenly spaced numbers between 0 and 2*pi
y = np.sin(x)  # plot sine curve for each value of x
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
# plt.plot(x,y)
```

```python
# What is matplotlib made of?
def get_attrs(obj):
  all_attrs = dir(obj)
  attrs = [x for x in all_attrs if x[0] != '_' and x[0].islower()]
  print(f'The object has {len(attrs)} attributes as follows:\n{attrs}')

def list_attrs(obj):
  all_attrs = dir(obj)
  attrs = [x for x in all_attrs if x[0] != '_' and x[0].islower()]
  return attrs
```

### Explanations

Backends: 
- Matplotlib has various use cases and output formats. 
- e.g. Python shell: pop up outputs, Jupyter notebook: inline plots, GUIs like PyQt, batchscripts: generate postscript images, web servers: dynamically serve plots. 
- Hence Matplotlib can target different outputs, each of these capabilities is called a backend. 

Selecting a backend (3 ways):
  - the rcParams["backend"] param in your matplotlibrc file
  - the MPLBACKEND environment variable
  - the matplotlib.use() function

Interactive figures: 
- pyplot module provides functions for explicitly creating figures that include interactive tools, a toolbar, a tool-tip and key bindings
- pyplot.figure : creates a figure.Figure or selects an existing figure
- pyplot.subplots : creates a figure.Figure and fills it with a grid of axes.Axes
- below is an example of creating and modifying a figure. It displays in a QtAgg GUI window. (NOTE: it WON'T WORK on Colab but will on Jupyter notebook. Have a look at Dropbox/Matplotlib/Matplotlib_General.ipynb (plot will pop up if backend = 'MacOS', ))

# Documenting all attributes

```python
def document_attrs(obj):
  methods = list_attrs(obj)
  docs = []
  for method in methods:
    docstring = getattr(obj, method).__doc__
    if docstring:
      line1 = docstring.split('\n')[0] or docstring.split('\n')[1] or docstring.split('\n')[2]
      line1 = line1.strip()
      docs.append((method, line1))
  return docs

def save_docs(obj):
  docs = document_attrs(obj)
  df = pd.DataFrame(docs)
  df.to_csv('quiz.csv', index=False)
```

```python
save_docs(ax)
```
