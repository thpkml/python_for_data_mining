```python
# @title Plot 2 lines in 1 graph
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# @title Add label/legend to the lines
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.legend()  # labels won't show without legend()
plt.show()
```

```python
# @title Change marker shape, size and color
plt.scatter(x, y, marker='X', color='r', s=100)
plt.show()
```

```python
# @title Create bar chart
x = ['A', 'B', 'C', 'D', 'E']
y = [4, 7, 2, 5, 8]

plt.bar(x, y)
plt.show()
```

```python
# @title Add title and axis labels
plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My Bar Chart')
plt.show()
```

```python
# @title Create stacked bar chart
x = ['A', 'B', 'C', 'D', 'E']
y1 = [4, 7, 2, 5, 8]
y2 = [1, 3, 5, 7, 9]
y3 = [1, 3, 5, 7, 9]
y4 = [1, 3, 5, 7, 9]
y5 = [1, 3, 5, 7, 9]

plt.bar(x, y1, label='y1')
plt.bar(x, y2, label='y2')
plt.legend()
plt.show()
# shows both bar overlapped (not stacked)
```

```python
plt.bar(x, y1, label='y1')
plt.bar(x, y2, bottom=y1, label='y2')  # this will stack y2 over y1
plt.bar(x, y3, bottom=y2, label='y3')  # this will stack y2 over y1
plt.bar(x, y4, bottom=y3, label='y4')  # this will stack y2 over y1
plt.bar(x, y5, bottom=y4, label='y5')  # this will stack y2 over y1
plt.legend()
plt.show()
# shows both bar overlapped (not stacked)
```

```python
# @title Create a histogram
data = [1, 1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9, 9, 10]
plt.hist(data)
plt.show()
```

```python
# @title Change histogram bins and colors
plt.hist(data, bins=4, color='g', edgecolor='w')
plt.show()
```

```python
# @title Get the bins
hist, bins, _ = plt.hist(data, bins=5, edgecolor='k')
print(bins)
plt.show()
```

```python
# @title Align histogram bins and xticks 
# xticks and bin edges don't align above
fig, ax = plt.subplots()
hist, bins, _ = ax.hist(data, bins=5, edgecolor='k')
ax.set_xticks(bins)
print(bins)
plt.show()
```

```python
# Question 16: scatter
x = list(range(11))
y = [xi**2 for xi in x]

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.title('Square of X')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# @title Scatter trend line
import numpy as np

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

plt.scatter(x, y, color='red', s=50)
plt.plot(x, p(x), color='blue')
plt.title('Square of X with Trend Line')
plt.xlabel('X')
plt.ylabel
```

```python
# @title Error bars - bar chart
x = ['A', 'B', 'C', 'D', 'E']
y = [4, 7, 2, 5, 8]
errors = [0.5, 1, 0.5, 0.8, 0.7]

fig, ax = plt.subplots()

# create the bar chart
ax.bar(x, y, color='purple')

# add error bars
ax.errorbar(x, y, yerr=errors, linestyle='None', color='black', capsize=3)

# set the title and axis labels
ax.set_title('Bar chart with error bars')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

plt.show()
```

```python
# @title Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 10, 40]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
# @title Pie slice explode
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 10, 40]
explode = [0, 0, 0, 0.1]  # explode slice 'D'

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')

plt.show()
```

```python
# @title Pie chart legend
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%')
ax.set_title('Pie Chart')

plt.legend(loc='best', labels=labels)  # loc: center, upper right, upper left etc.
plt.show()
```

```python
# @title 3D scatter

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [3, 6, 9, 12, 15]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# @title Multiple subplots in a figure

# create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# plot data on the first subplot
ax1.plot(x, y1, 'r-', label='sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Plot 1')
ax1.legend()

# plot data on the second subplot
ax2.plot(x, y2, 'b-', label='cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Plot 2')
ax2.legend()

# adjust the spacing between the subplots
plt.subplots_adjust(hspace=0.5)

# show the plot
plt.show()
```

```python
# @title Grid of plots

# Generate some random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2)

# Draw each plot on its own subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sin(x)')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cos(x)')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tan(x)')
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exp(x)')

# Add a title for the whole figure
fig.suptitle('Four plots in a 2x2 grid')

# Display the plot
plt.show()
```

```python
# bigger figure size
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

# Draw each plot on its own subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sin(x)')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cos(x)')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Tan(x)')
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exp(x)')

# Add a title for the whole figure
fig.suptitle('Four plots in a 2x2 grid')

# Display the plot
plt.show()
```
