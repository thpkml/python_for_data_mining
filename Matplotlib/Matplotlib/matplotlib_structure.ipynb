```python
# @title Artist Layer
import numpy as np
import matplotlib.pyplot as plt

# create some data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# create a Figure and Axes object
fig, ax = plt.subplots()

# create a Line2D object to represent the line
line = plt.Line2D(x, y, linewidth=2, markersize=8, linestyle='-', marker='o')

# add the line to the Axes
ax.add_line(line)

# change the line color and marker style
line.set_color('red')
line.set_marker('s')

# add text to the plot using a Text object
text = plt.Text(2, 0, 'sin(x)', fontsize=14, color='blue')
ax.add_artist(text)

# set the x and y limits of the Axes
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1, 1)

# show the plot
plt.show()
```

```python
# @title Scripting Layer
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis object
plt.figure()  # gives you a blank canvas where you can add your plot elements

# Plot the data
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine wave')

# Show the plot
plt.show()
```

```python
# @title Customisation using Scripting layer
import numpy as np
import matplotlib.pyplot as plt

# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure and axes objects
fig, ax = plt.subplots(figsize=(8, 6))

# Plot sin and cos curves with custom line styles and colors
ax.plot(x, y1, linestyle='-', linewidth=2, color='red', label='sin(x)')
ax.plot(x, y2, linestyle='--', linewidth=2, color='blue', label='cos(x)')

# Add title and axis labels
ax.set_title('Sin and Cos Curves')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Customize x and y tick labels and tick locations
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_xticklabels(['0', '2π/5', '4π/5', '6π/5', '8π/5', '2π'])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.set_yticklabels(['-1', '-0.5', '0', '0.5', '1'])

# Customize legend
ax.legend(loc='upper right')

# Show plot
plt.show()
```

```python
# @title Scripting layer vs Artist layer
# how the use of artist layer gives more control
# scripting layer code
import matplotlib.pyplot as plt
import numpy as np

# create some data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# create the plot
plt.plot(x, y)

# customize the x-axis tick labels
xticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xticklabels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(xticks, xticklabels)

# show the plot
plt.show()
```

```python
# artist layer
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# create some data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# create the plot
fig, ax = plt.subplots()
ax.plot(x, y)

# customize the x-axis tick labels using the artist layer
xticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xticklabels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)

# use a custom formatter to add a unit to the last tick label
def format_last_tick(x, pos):
    if x == xticks[-1]:
        return f'{x}$^\circ$'
    else:
        return str(x)

ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_last_tick))

# show the plot
plt.show()
```
