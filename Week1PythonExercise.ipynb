```python
# @title Q 5
for i in range(10):
  print('*'*i)
```

```python
# more advanced
def pyramid(height):
    for i in range(1, height + 1):
        spaces = height - i
        asterisks = 2 * i - 1
        print(' ' * spaces + '*' * asterisks)

height = int(input("Enter the height of the pyramid: "))
pyramid(height)
```

```python
# @title Q 6
def circle_area(radius):
    """
    This function calculates the area of a circle given its radius.
    """
    return 3.14 * radius * radius

radius = float(input("Enter the radius of the circle: "))
area = circle_area(radius)
print(f"The area of the circle is {area}")
```

```python
# @title Q 7 
def sum_of_numbers(numbers):
    """
    This function takes a list of numbers and returns the sum of all numbers in the list.
    """
    # Initialize a variable to store the sum
    total = 0
    
    # Loop over each number in the list
    for num in numbers:
        # Add the current number to the running total
        total += num
    
    # Return the sum of all numbers
    return total

# Create a list of numbers
numbers = [1, 2, 3, 4, 5]

# Call the function and store the result in a variable
result = sum_of_numbers(numbers)

# Print the result
print(f"The sum of all numbers in the list is {result}")
```

```python
# @title Q 14
# Palindrome
def is_palindrome(s):
    """
    This function checks if a string is a palindrome.
    """
    # Convert the string to lowercase and remove any spaces
    s = s.lower().replace(" ", "")
    
    # Check if the string is equal to its reverse
    return s == s[::-1]

# Get the string from the user
s = input("Enter a string: ")

# Call the function and store the result in a variable
result = is_palindrome(s)

# Print the result
if result:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
# @title Q 20
# Greatest Common Divisor
def gcd(a, b):
    """
    This function calculates the greatest common divisor (GCD) of two numbers.
    """
    # Check if either number is 0, and return the other number if so
    if a == 0:
        return b
    if b == 0:
        return a

    # Keep finding the remainder until one of the numbers is 0
    while b != 0:
        a, b = b, a % b
    
    # Return the non-zero number, which is the GCD
    return a

# Get the two numbers from the user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

# Call the function and store the result in a variable
result = gcd(a, b)

# Print the result
print(f"The GCD of {a} and {b} is: {result}")
```
