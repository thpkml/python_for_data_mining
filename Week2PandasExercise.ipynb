```python
# @title sub, mean, abs
# Create a DataFrame and use a method to return a new DataFrame with the absolute difference between each value and the mean.
# Create a DataFrame with random values
df = pd.DataFrame(np.random.randn(5, 3), columns=['A', 'B', 'C'])

# Get the mean of the DataFrame
mean = df.mean()

print(df, '\n')
print('mean\n', mean, '\n')
# Use the sub method to subtract the mean from each value and return the absolute value
abs_diff = df.sub(mean).abs()

print(abs_diff)
```

```python
# @title interpolate
# Create a DataFrame with some missing values and use a method to interpolate the missing values using linear interpolation.

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [2, 4, 6, np.nan, 10],
    'C': [5, 6, 7, 8, 9]
})

# Interpolate the missing values using linear interpolation
df_interpolated = df.interpolate()

# Print the original DataFrame
print("Original DataFrame:\n", df)

# Print the interpolated DataFrame
print("\nInterpolated DataFrame:\n", df_interpolated)
```

How does interpolation work?  
Linear (default):  
missing value lies in the straight line connecting the existing adjacient values

Quadratic:  
Quadratic interpolation is a method of estimating the value of a function based on three neighboring data points. It involves fitting a quadratic function to the three nearest data points and using this function to interpolate the value of the function at a desired point. 

Polynomial:  
e.g. method='polynomial', order=2

```python
# Create a DataFrame with some missing values
df = pd.DataFrame({'A': [3, np.nan, 3.5, 2.4, np.nan, 9, 6.4, 8, 7]})
print(df)

# Interpolate missing values using linear interpolation
df_interpolated = df.interpolate(method='linear')
print(df_interpolated)

# Interpolate missing values using quadratic interpolation
df_interpolated = df.interpolate(method='quadratic')
print(df_interpolated)

# Interpolate missing values using polynomial interpolation with order=2
df_interpolated = df.interpolate(method='polynomial', order=3)
print(df_interpolated)
```

```python
# @title groupby, get_group
# Create a DataFrame and use a method to return a new DataFrame with the rows grouped by a specific column.
# Create a DataFrame
data = {'Animal': ['Dog', 'Cat', 'Dog', 'Cat', 'Dog', 'Cat'],
        'Weight': [25, 10, 15, 8, 22, 12],
        'Height': [60, 30, 45, 20, 55, 25]}
df = pd.DataFrame(data)

# Group the DataFrame by the 'Animal' column
grouped = df.groupby('Animal')

# Get the groups as separate DataFrames
dog_group = grouped.get_group('Dog')
cat_group = grouped.get_group('Cat')

print(df, '\n')
print(dog_group, '\n')
print(cat_group)
```

```python
df.groupby('Animal').agg({'Weight':'sum', 'Height':'mean'})
```
