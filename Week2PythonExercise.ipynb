```python
# @title For Loop
# Write a Python function that takes an integer as input and prints a countdown 
# ...from that number to 0 using a for loop and range.
def countdown(num):
    for i in range(num, -1, -1):
        print(i)

countdown(4)
```

```python
# Write a Python function that takes a list of strings as input and prints each 
# ...string in the list with its index using a for loop and range.
def print_strings_with_index(str_list):
    for i in range(len(str_list)):
        print(i, str_list[i])

my_str_list = ['dancing', 'in', 'the', 'rain']
print_strings_with_index(my_str_list)
```

```python
# @title While Loop
# Write a Python function that takes an integer as input and prints the sum of 
# ...all the integers from 0 to that number using a while loop.
def sum_integers(num):
    i = 0
    total = 0
    while i <= num:
        total += i
        i += 1
    print(total)

sum_integers(4)
```

```python
# Write a Python function that takes a list of strings as input and prints each 
# ...string in the list until it reaches a string that starts with the letter 'z' using a while loop.
def print_until_z(str_list):
    i = 0
    while i < len(str_list) and not str_list[i].startswith('z'):
        print(str_list[i])
        i += 1

my_str_list2 = 'A battle between zombies and aliens'.split(' ')

print(my_str_list2)

print_until_z(my_str_list2)
```

```python
# @title Conditionals
# Write a Python function that takes a list of integers as input and returns the 
# ...sum of all the positive numbers and the product of all the negative numbers.
def sum_and_product(num_list):
    sum_positive = 0
    product_negative = 1
    for num in num_list:
        if num > 0:
            sum_positive += num
        elif num < 0:
            product_negative *= num
    return sum_positive, product_negative

my_int_list = [-2, -3, -4, 2, 3, 4]

sum_and_product(my_int_list)
```

```python
# Write a Python function that takes a list of strings as input and returns a new 
# ...list containing only the strings that have a length greater than 5 or start with the letter 'a'.
def filter_strings(str_list):
    filtered_list = []
    for string in str_list:
        if len(string) > 5 or string.startswith('a'):
            filtered_list.append(string)
    return filtered_list

filter_strings(my_str_list2)
```

```python
# @title Functions
# Write a Python function that takes two numbers as input and returns the sum, 
# ...difference, product, and quotient of the two numbers.
def arithmetic_operations(num1, num2):
    sum_num = num1 + num2
    difference = num1 - num2
    product = num1 * num2
    quotient = num1 / num2
    return sum_num, difference, product, quotient

arithmetic_operations(3, 4)
```

```python
# Write a Python function that takes a list of strings as input and returns a new 
# ...list with all the vowels removed from each string.
def remove_vowels(str_list):
    vowels = 'aeiouAEIOU'
    new_list = []
    for string in str_list:
        new_string = ''
        for letter in string:
            if letter not in vowels:
                new_string += letter
        new_list.append(new_string)
    return new_list

remove_vowels(my_str_list)
```
