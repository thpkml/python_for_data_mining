```python
# @title concat
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   #index=[0, 1, 2, 3])
                   index=[4, 5, 6, 7])
result = pd.concat([df1, df2], axis=1)
result
```

```python
# @title crosstab
# creates a cross-tabulation table. It's used to summarize the relationship between two or more categorical variables.
data = {'fruit': ['apple', 'banana', 'banana', 'apple', 'apple'],
        'color': ['red', 'yellow', 'yellow', 'red', 'green']}
df = pd.DataFrame(data)

# create a cross-tabulation table
pd.crosstab(df['fruit'], df['color'])
```

```python
# @title cut
# discretize continuous data into intervalled bins
np.random.seed(0)
data = np.random.rand(5)*100
df = pd.DataFrame({'data': data})
# define bin intervals
bins = [0,25,50,75,100]
# descretize data into bins
df['binned'] = pd.cut(df['data'], bins)
df
```

```python
# @title melt
# reshape DF by 'melting' 1 or more columns into new columns
df = pd.DataFrame({
    "Person": ["John", "Jane", "Jim", "Joan"],
    "Year": [2020, 2021, 2022, 2021],
    "Score1": [90, 80, 85, 95],
    "Score2": [95, 85, 90, 80]
})

df.melt(id_vars=["Person", "Year"], value_vars=["Score1", "Score2"])
```

```python
# @title merge
# combine 2 or more DFs based on 1 or more columns
df1 = pd.DataFrame({
    "key": [1, 2, 3, 4],
    "value1": ["A", "B", "C", "D"]
})

df2 = pd.DataFrame({
    "key": [3, 4, 5, 6],
    "value2": ["W", "X", "Y", "Z"]
})

pd.merge(df1, df2, on="key")
```

```python
# @title merge_asof
# 'as-of' join between 2 DFs
pd.merge_asof(df1, df2, on='key')
```

```python
# @title merge_ordered
# perform an ordered join between two DataFrames
pd.merge_ordered(df1, df2, on='key')
```

```python
# @title to_timedelta
# convert string to Timedelta
timedelta = pd.to_timedelta("5 days")
print(timedelta)
# Output: 5 days 00:00:00

# convert integer to Timedelta
timedelta = pd.to_timedelta(5, unit='D')
print(timedelta)
# Output: 5 days 00:00:00

# convert float to Timedelta
timedelta = pd.to_timedelta(5.5, unit='D')
print(timedelta)
# Output: 5 days 12:00:00
```

```python
# @title wide_to_long
# reshape df from wide to long format
# wide format: multiple columns with same data but differentiated by some column suffix
# long format: all the data in a single column. An additional column to distinguish different columns

# create a sample dataframe in wide format
df = pd.DataFrame({
    "id": [1, 2, 3],
    "var_A_1": [10, 20, 30],
    "var_A_2": [100, 200, 300],
    "var_B_1": [1000, 2000, 3000],
    "var_B_2": [10000, 20000, 30000],
})

# reshape from wide format to long format
df_long = pd.wide_to_long(df, ["var_A", "var_B"], i="id", j="time_period")

# show resulting long format
print(df_long)
```
