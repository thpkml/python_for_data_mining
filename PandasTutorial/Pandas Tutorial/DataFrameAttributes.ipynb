### Using AND OR in filters

```
filt = (df['pcm'] > 1500) & (df['pcm'] < 2500) & (df['bedrooms'] > 4)
so_cheap = df.loc[filt, ['pcm', 'bedrooms', 'postcode', 'station', 'telephone']]
so_cheap.head(10)
```

```
# Show pcm column for all listings that DO NOT MATCH the filter
df.loc[~filt, 'pcm']
```

#### applymap(function_name) - for the whole Dataframe

```
# to get len of each value in the df
# df.applymap(len)
# but it cannot be applied if there in non string values in the df 
# Need to write a custom function (CHECK BELOW)
```

```
# df.applymap(str.lower) # this would convert all string to lower if there were no int fields
# Needs a custom function as follows
def map_lower(value):
  if type(value) == str:
    return value.lower()
  else:
    return value

def map_len_all(value):
  return len(str(value)) # convert to str if non str and get len

def map_len_str(value):
  if type(value) == str:
    return len(value)
  else:
    return value # only return len of str, if no str return original value
```

```
df.applymap(map_lower).head()
```

```
# check len of all values
df.applymap(map_len_all).head()
```

```
# check len of string values, return non string as is
df.applymap(map_len_str).head()
```

### Expanding a Column into Multiple Columns

```
# separate 'address' at the ',' 
df['address'].str.split(',') # to just split it into a list
```

```
# to expand the split values into separate columns: 
df['address'].str.split(', ', expand=True)
```

```
# UH OH!! there are up to 10 separate words in some addresses
# might need to use custom function
def split_address(value):
  words = value.split(', ')
  return words[0] + ', ' + words[-1] # the street and the city(last words)

# use the function
df['address'].apply(split_address)
```

```
# persist the change in place
df['address'] = df['address'].apply(split_address)
# Expand the address to two columns
df['address'].str.split(', ', expand=True)
```

```
# persist the expanded columns with new names
df[['street', 'city']] = df['address'].str.split(', ', expand=True)
df
# scroll to the end to find 'street' and 'city' columns
```

```
# bringing things to view
df[['address', 'city', 'street', 'postcode', 'pcm']]
```
